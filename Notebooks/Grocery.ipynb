```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from apyori import apriori
```

```python
ds = pd.read_csv("Datasets/grocery.csv")
ds.head()
```

```python
ds.shape
```

```python
ds.drop(['Item(s)'],  axis=1)
```

```python
groceries = []
for i in range(1, 7501):
    groceries.append([str(ds.values[i, j]) for j in range(0, 20)])
```

```python
association_rules = apriori(groceries, min_support=0.0045, min_confidence=0.2, min_lift=3, min_length=2)
association_results = list(association_rules)
```

```python
print(f'There are {len(association_results)} Relation derived.')
```

```python
for i in range(0, len(association_results)):
    print(association_results[i][0])
```

```python
for item in association_results:
    pair = item[0]
    items = [x for x in pair]
    print("Rule: " + items[0] + " -> " + items[1])

    print("Support: " + str(item[1]))

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```
