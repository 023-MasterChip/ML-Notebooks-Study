```python
ds = dataset
ds.head()
```

```python
ds = ds.drop(['PassengerId', 'Name', 'Ticket', 'Fare', 'Cabin', 'Embarked' ], axis=1)
```

```python
ds.head()
```

```python
ds.groupby('Survived').size()
```

```python
# Resample
from sklearn.utils import resample
#create two different dataframe of majority and minority class
ds_departed = ds[(ds['Survived']==0)]
ds_survived = ds[(ds['Survived']==1)]

# upsample minority class
ds_survived_upsampled = resample(ds_survived,
                                 replace=True,    # sample with replacement
                                 n_samples= 815, # to match majority class
                                 random_state=42)  # reproducible results

# Combine majority class with upsampled minority class
new_ds = pd.concat([ds_departed, ds_survived_upsampled])
```

```python
new_ds.head()
```

```python
new_ds.groupby('Survived').size()
```

```python
sns.countplot(x = 'Survived', data=new_ds, order = dataset['Survived'].value_counts().index)
```

```python
new_ds.info()
```

```python
sex = {"male":0,'female':1}
new_ds['Sex'] = dataset['Sex'].map(sex)
```

```python
survive = {0:'Perish',1:'Survive'}
new_ds['Survived'] = dataset['Survived'].map(survive)
```

```python
new_ds
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.cluster import KMeans

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import accuracy_score, classification_report, silhouette_score
```

```python
x = new_ds.iloc[:, 1:6].values
y = new_ds.iloc[:, 0].values
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 1)
```

```python
print("x_train: ",x_train.shape)
print("y_train: ",y_train.shape)
print("x_test: ",x_test.shape)
print("y_test: ",y_test.shape)
```

```python
model_lr = LogisticRegression(max_iter = 600)
model_lr.fit(x_train, y_train)
```

```python
y_pred_lr = model_lr.predict(x_test)
print("Accuracy : ", accuracy_score(y_test, y_pred_lr))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model_knn = KNeighborsClassifier()
model_knn.fit(x_train,y_train)
```

```python
y_pred_knn = model_knn.predict(x_test)
print("Accuracy : ", accuracy_score(y_test, y_pred_knn))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_dt = DecisionTreeClassifier()
model_dt.fit(x_train,y_train)
```

```python
y_pred_dt = model_dt.predict(x_test)
print("Accuracy : ", accuracy_score(y_test, y_pred_dt))
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model_nb = GaussianNB()
model_nb.fit(x_train,y_train)
```

```python
y_pred_nb = model_nb.predict(x_test)
print("Accuracy : ", accuracy_score(y_test, y_pred_nb))
```

```python
from sklearn.svm import SVC
```

```python
model_svc = SVC(gamma='auto')
model_svc.fit(x_train,y_train)
```

```python
y_pred_svc = model_svc.predict(x_test)
print("Accuracy : ", accuracy_score(y_test, y_pred_svc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rf = RandomForestClassifier(n_estimators = 30, criterion = 'entropy', random_state = 0)
model_rf.fit(x_train,y_train)
```
